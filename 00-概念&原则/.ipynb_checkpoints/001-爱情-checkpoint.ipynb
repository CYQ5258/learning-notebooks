***什么是爱情？***
- 爱情是为了 追求/体验 不同的、美好的、幸福的生活，而不是为了 `忍辱负重`（这里具体指由于脾气/··· 不和/···，彼此频繁受气；而不是一起度过贫寒这种）

***什么是现实？***
- 在我看来，婚姻之前的都可以称之为爱情；
    - 然而，当涉及到婚姻时，那么就必须考虑`现实`（婚姻代表着家庭，代表着责任，对彼此的责任，对孩子的责任，对父母的责任，
        - 我作为一个男人，更要肩负起这些责任
    - 当然，如果最初的目的是要追求一段`不求天长地久, 只求曾经拥有`的爱情，那么就不要去理会什么`现实`，**但是**当对一段感情上升到`婚姻`的高度时，就必须要分析各种因素，因为每一个人的`时间和精力`是有限的，选择了这一个，那么很可能就会错过另一个，无论如何，我们是要为我们自己当初的`选择`负责的
    

# 亲密

## 彼此了解
- 首先，要和她接触，彼此要进行最基本的了解（信息交换）
    - 不要仅仅从表面去判断一个人，或许你并没有看到`真正的她/他`
    - 当然，这样也是需要时间和精力成本的，所以所谓的 **`第一印象`** 很重要
        - 对于男生来说：第一印象的好坏很大程度取决于女生的长相、气质
        - 对于女生来说：emmmm
    - 性（我不会去避讳这个概念）也同样是影响因素，是激素引起的生理上的本能反应，无法避免。但是，在我看来，性行为不是目的，而是增进感情的行为/过程

## 朋友做起
- 在此基础上，如果 **`彼此`** 感觉对方是符合自己 `胃口` 的人，那么就从朋友做起，彼此之间进行更深的了解
    - 当然，我自己学过一些两性吸引学相关的一些知识，有一些 `技巧` 可以加快这个过程
    - `'真正的感情'` 是彼此长时间呆在一起，从而形成的依恋/依赖感
- 不用着急确定关系（呃。。。特殊目的除外），前期可以像朋友一样多在一起走动，试着去融入彼此的生活，尽可能的去判断彼此是适合爱情、还是适合友情。
    - 爱情是需要付出更多的时间和精力的，并且，如果刚开始就确定了关系，最终发现不合适，不仅付出的感情付水东流（当然，如果要说从中可以汲取教训的话，那也无可厚非）

# 承诺

- 做彼此的 **`避风港`**
- 相互包容，共同成长
    - 人总是会犯错，但是发现自己犯了错，首先要马上用嘴道歉；随后要开始用行动道歉、弥补，直至一切恢复原状，甚至更好（可能需要付出很多代价，但是这是必须做的！）
    - 处于社会底层阶段的彼此，或许很多方面都不够好，需要很长时间去学习和提升（当然，前提是彼此都要有一颗上进的心，彼此相互鼓励）

# 原则

## 自我坚守

- 人总是会 **`犯错`** ，但是发现自己犯了错，首先要马上 **`用嘴道歉`** ；随后要开始 **`用行动道歉`** 、弥补，直至一切恢复原状，甚至更好（可能需要付出很多代价，但是这是必须做的！）
- 如果我选择了她，那么我会 `无条件的选择相信她` ，不去影响她的私人生活；但是一旦让我知道了超出原则的事情，不会再有第二次机会
- 如果我选择了她，她也选择了我，那么，或许某些时候我会对她 `说谎` ，但是，同样，这个说谎的前提只能想考试作弊一样， **`不会对她造成伤害`**

## 决定在一起？

- 外表不用很出众（当然，外表好看会加分），只要彼此看着顺眼就行
    - 即使外表很好看，然而随着在一起的时间的延长，也会逐渐淡化
- 以结婚为前提的爱情 的 `最终目的是婚姻` ，所以 `结婚前的爱情` （在彼此眼中）都属于 `“预选”` 阶段，觉得彼此不合适要先和对方 `沟通清楚` ，要做到 `当断则断` （当然，无论当时做出什么样的 `选择` ，之后都必须要肩负起 `责任` ）
- 有些问题只有在一起了才能发现要怎么办？
    - 首先，相互沟通，彼此监督、进步，同时变得更优秀（前提是彼此都有自学能力，都有上进心）
    - 如果对方拒绝，并且已经超出了自己的原则，那么当断则断

## 是否退出？

- 当发觉双方在一起的确 **`不合适`** （这个概念目前很模糊，需要你自己去发觉）的时候，我会选择主动退出（当然这是当前自己的想法，最终是否能执行又是另一回事）
    - 不合适？
        - 比如当真正走向社会时，双方产生不同的观点，彼此有不同的选择
        - 各自的 **`价值观`** 不同
            - 价值观：什么是对的；什么最重要；什么更重要；什么一定不能做；什么事一定要做到底。
